# Generate Classification Data With Sklearn

As a data scientist, you will eventually create a machine learning model. To test your model you need data with different characteristics for a complete overview of the model's performance. Because real-life data is limited, you must know how to generate data with different characteristics. Scikit-learn offers us extremely useful tools for the generation of a synthetic dataset for ML. See bellow.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import seaborn as sns; sns.set(color_codes=True)
# Generate Clean data

X,y = make_classification(n_samples=100, n_features=2, n_redundant=0, n_repeated=0, n_classes=2, n_clusters_per_class=1,class_sep=2,flip_y=0,weights=[0.5,0.5])
f, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(20,8))
sns.scatterplot(X[:,0],X[:,1],hue=y,ax=ax1);
ax1.set_title("No Noise");


# Generate noisy Data
X,y = make_classification(n_samples=100, n_features=2,  n_redundant=0, n_repeated=0, n_classes=2, n_clusters_per_class=1,class_sep=2,flip_y=0.2,weights=[0.5,0.5])
sns.scatterplot(X[:,0],X[:,1],hue=y,ax=ax2);
ax2.set_title("With Noise");
plt.show();
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from itertools import combinations
```

### Project Overview
For this project I used Random Forest ( Assemble of Decision Trees), to predict Sale Price of Iowa Houses. To make the model explainable, I performed feature selection for numerical columns with the correlation matrix method and the built-in feature importance attribute from the Random forest model.
I was able to use fewer variables and obtain better results. I used simple imputation to deal with missing values of numerical columns and label encoder for categorical data.
Additionally, I experimented with feature interaction and obtained the last result.

### Feature selection
#### Feature Importance

Feature importance gives you a score for each feature of your data, the higher the score more important or relevant is the feature towards your output variable. I have used the built in feature importance method for Random forest model.

```python
data = pd.read_csv('data/train.csv', index_col = "Id")
categorical_cols = data.select_dtypes(include=['object']).columns
numerical_cols = data.select_dtypes(exclude=['object']).columns
data_numerical = data[numerical_cols]
# Prepare data for training and validation
temp_data = data_numerical.drop(["LotFrontage","GarageYrBlt"], axis = 1)
temp_data = temp_data.dropna(axis = 0)

y = temp_data.SalePrice
X = temp_data.drop("SalePrice",axis =1)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.20, random_state=1)
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train, y_train)
preds = model.predict(X_valid)
print(mean_absolute_error(y_valid, preds))

feat_importances = pd.Series(model.feature_importances_, index=X.columns)


feat_importances.nlargest(15).plot(kind='barh')
plt.grid()
plt.show()
```

#### Correlation Matrix

Correlation states how the features are related to each other or the target variable.
The correlation can be positive or negative, for this dataset are the variables highlly correlated are also positively correlated.

`The Heatmap bellow makes it easy to identify which features are most related to the target variable.`

```python
import seaborn as sns

corrmat = temp_data.corr()
cor = list(corrmat["SalePrice"].sort_values(ascending=False).head(15).index)
cor.pop(0)

top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
col_set = ['OverallQual','GarageArea','TotalBsmtSF','YearRemodAdd','WoodDeckSF','Fireplaces','OpenPorchSF','OverallCond','FullBath','LotArea','BsmtFinSF1','GrLivArea','2ndFlrSF','MasVnrArea','TotRmsAbvGrd','YearBuilt','GarageCars','1stFlrSF']
feat_imp = ['OverallQual','GrLivArea','TotalBsmtSF','1stFlrSF','BsmtFinSF1','2ndFlrSF',
 'GarageArea','GarageCars','YearBuilt','LotArea','YearRemodAdd','MasVnrArea','OpenPorchSF','WoodDeckSF','OverallCond']
cor_mat = ['OverallQual','GrLivArea','GarageCars','GarageArea','TotalBsmtSF','1stFlrSF','FullBath',
 'TotRmsAbvGrd','YearBuilt','YearRemodAdd','MasVnrArea','Fireplaces','BsmtFinSF1','WoodDeckSF']
```

### Result of using feature selection:

```python
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train, y_train)
preds = model.predict(X_valid)
print("MAE with all numerical Columns:",round(mean_absolute_error(y_valid, preds),2))

model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train[list(feat_importances.nlargest(15).index)], y_train)
preds = model.predict(X_valid[list(feat_importances.nlargest(15).index)])
print("MAE with top 15 Feature Importance Columns:", round(mean_absolute_error(y_valid, preds),2))


model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train[cor_mat], y_train)
preds = model.predict(X_valid[cor_mat])
print("MAE with top 15 Correlated Columns:", round(mean_absolute_error(y_valid, preds),2))

model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train[col_set], y_train)
preds = model.predict(X_valid[col_set])
print("MAE the set of 18 Columns from the Two Method of feature selection:", round(mean_absolute_error(y_valid, preds),2))
```

We can see that the model with only the Feature Importance variables performed better. But the model was only validated once. With each different train and validation set, the results may slightly differ. 
To keep more columns for experimenting ( and because the results are similar for the Model with all numerical columns and the one with only 18 columns - the set obtained from both methods), I will work with the set of 18 columns. At this point, this is very good because we have made the model more explainable and more simple with virtually the same results.

### Train Model With Numerical Data

Now that I have select which numerical columns to use, I decided to use simple imputation to deal with missing values. For each value imputed I added a corresponding boolean collomn that Idetifies if the value was missing. This small change improved the model and we obtained a smaller MAE.

The only column with missing values was `MasVnrArea`, therefore, we added only one boolean column. That is , we have not increased the size of the dataset too much.

```python
data = pd.read_csv('data/train.csv', index_col = "Id")
numerical_cols = data.select_dtypes(exclude=['object']).columns

# Prepare data for training and validation
y = data.SalePrice
X = data.drop("SalePrice",axis =1)

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
missing_col_series =data[col_set].isnull().sum()[data[col_set].isnull().sum() > 0]
missing = missing_col_series.index

# Make copy to avoid changing original data (when imputing)
X_train_plus = X_train[col_set].copy()
X_valid_plus = X_valid[col_set].copy()

# Make new columns indicating what will be imputed
for col in missing:
    X_train_plus[col + '_was_missing'] = X_train_plus[col].isnull()
    X_valid_plus[col + '_was_missing'] = X_valid_plus[col].isnull()

# Imputation
my_imputer = SimpleImputer()
imputed_X_train_plus = pd.DataFrame(my_imputer.fit_transform(X_train_plus))
imputed_X_valid_plus = pd.DataFrame(my_imputer.transform(X_valid_plus))

# Imputation removed column names; put them back
imputed_X_train_plus.columns = X_train_plus.columns
imputed_X_valid_plus.columns = X_valid_plus.columns

# Imputation removed column names; put them back
imputed_X_train_plus.index = X_train_plus.index
imputed_X_valid_plus.index = X_valid_plus.index

model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(imputed_X_train_plus, y_train) 
preds = model.predict(imputed_X_valid_plus)
print("MasVnrArea had 8 missing values ",sum(imputed_X_train_plus['MasVnrArea_was_missing']))
print("MAE for imputed numerical columns :", round(mean_absolute_error(y_valid, preds),2))
```

### Train Model with Categorical and Numerical Data

For using categorical variables, I selected the columns that have the same set of values in both the train and validation set. This was done so that the label encoder can word properly. 

 I had to remove a few extra columns due to value errors. The columns were removed for practical reasons and you can find them saved on the problem_cols variable. Besides that, other columns were not considered because the training and validation set for the columns were different, which causes a problem with the label encoder.

` Adding categorical columns to the model, reduced the MAE from 16200 to 15807. `


```python
categorical_cols = data.select_dtypes(include=['object']).columns

# Columns that can be safely label encoded
good_label_cols = [col for col in categorical_cols if 
                   set(X_train[col]) == set(X_valid[col])]

label_encoder = LabelEncoder()

temp = ['MSZoning','Street','LotShape','LandContour','LotConfig','LandSlope','BldgType','HouseStyle','ExterQual',
 'CentralAir','KitchenQual','PavedDrive','SaleCondition']

problem_cols = list(set(good_label_cols) - set(temp))

for col in temp:
    label_X_train[col] = label_encoder.fit_transform(X_train[col])
    label_X_valid[col] = label_encoder.transform(X_valid[col])

Last_X_train = imputed_X_train_plus.merge(label_X_train[temp], left_index=True, right_index=True)
Last_X_valid = imputed_X_valid_plus.merge(label_X_valid[temp], left_index=True, right_index=True)

model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(Last_X_train, y_train) 
preds = model.predict(Last_X_valid)
print("MAE for model with categorical variables:", round(mean_absolute_error(y_valid, preds),2))
                                                                         
```

### Feature Interaction

 I created a set of combinations from which to experiment. A few of the experiments are printed bellow. I was able to find three interactions that further reduced the Mean Absolute error.

The procedure is the following:
* I selected the top 5 features from the feature importance and correlation matrix methods.
* Them I create a list of all possible combinations.
* I tested how each interaction affects the model individually.
* Then I chose the top 5 individual interactions that produced an error bellow the previous model (without interactions).
* I experimented with different sets of the 5 top individual interactions and found that using three created the most reduced MAE.
    * OverallQual *  BsmtFinSF1
    * 1stFlrSF * TotalBsmtSF and
    * TotalBsmtSF * OverallQual



```python
feat = list(feat_importances.nlargest(15).index)
comb_col = list(set(feat[0:5] + cor[0:5]))
comb_all = []
for i in combinations(comb_col,2):
    comb_all.append(i)
comb_all 
```

```python
score = dict()
for i in comb_all:
    new_col = Last_X_train[i[0]] * Last_X_train[i[1]]
    Last_X_train["temp"] = new_col

    new_col = Last_X_valid[i[0]] * Last_X_valid[i[1]]
    Last_X_valid["temp"] = new_col

    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(Last_X_train, y_train) 
    preds = model.predict(Last_X_valid)
    score[i] = [mean_absolute_error(y_valid, preds)]
```

```python
pd.DataFrame(score).transpose().sort_values(by= 0).head(5)
```

```python
new_col = Last_X_train["OverallQual"] * Last_X_train["BsmtFinSF1"]
Last_X_train["temp"] = new_col
new_col = Last_X_train["1stFlrSF"] * Last_X_train["TotalBsmtSF"]
Last_X_train["temp2"] = new_col


new_col = Last_X_valid["OverallQual"] * Last_X_valid["BsmtFinSF1"]
Last_X_valid["temp"] = new_col
new_col = Last_X_valid["1stFlrSF"] * Last_X_valid["TotalBsmtSF"]
Last_X_valid["temp2"] = new_col


model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(Last_X_train, y_train) 
preds = model.predict(Last_X_valid)
mean_absolute_error(y_valid, preds)
```

```python
Last_X_train = imputed_X_train_plus.merge(label_X_train[temp], left_index=True, right_index=True)
Last_X_valid = imputed_X_valid_plus.merge(label_X_valid[temp], left_index=True, right_index=True)


new_col = Last_X_train["OverallQual"] * Last_X_train["BsmtFinSF1"]
Last_X_train["temp"] = new_col
new_col = Last_X_train["1stFlrSF"] * Last_X_train["TotalBsmtSF"]
Last_X_train["temp2"] = new_col
new_col = Last_X_train["TotalBsmtSF"] * Last_X_train["OverallQual"]
Last_X_train["temp3"] = new_col


new_col = Last_X_valid["OverallQual"] * Last_X_valid["BsmtFinSF1"]
Last_X_valid["temp"] = new_col
new_col = Last_X_valid["1stFlrSF"] * Last_X_valid["TotalBsmtSF"]
Last_X_valid["temp2"] = new_col
new_col = Last_X_valid["TotalBsmtSF"] * Last_X_valid["OverallQual"]
Last_X_valid["temp3"] = new_col



model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(Last_X_train, y_train) 
preds = model.predict(Last_X_valid)
print("MAE for model with Interactions:", round(mean_absolute_error(y_valid, preds),2))
          
```

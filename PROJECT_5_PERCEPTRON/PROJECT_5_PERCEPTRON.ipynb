### PCA

By performing dimensionality reduction with principal component analysis, we can see that the data is linearly separable.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
X_ = new_iris[["sepal_length","sepal_width","petal_length","petal_width"]].copy()
y_ = new_iris["species"]
X = StandardScaler().fit_transform(X_)
pca = PCA(n_components = 2)
principalcomp = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalcomp
             , columns = ['principal component 1', 'principal component 2'])
target_t = pd.DataFrame({'y':y_})
finalDf = pd.concat( [principalDf,target_t[['y']]], axis = 1)
finalDf.head(5)

plt.figure(figsize = (8,8))
sns.scatterplot(
    x='principal component 1',
    y='principal component 2',
    data=finalDf,    
    hue='y',
    legend= "full"
    
)

plt.title('2 Component PCA', fontsize = 15)
plt.grid()
# both, hue and size are optional
sns.despine() # prettier layout
```

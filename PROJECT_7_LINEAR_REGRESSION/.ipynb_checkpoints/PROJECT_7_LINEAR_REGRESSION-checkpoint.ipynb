## Predicting a Measure of Disease Progression for Diabetes Patients

**Ten baseline variables are used:**
* age 
* sex 
* body mass index
* average blood pressure
* six blood serum measurements

**About the study for which the data was obtained:**

Diabetes study: 442 diabetes patients were measured on 10 baseline variables; a prediction model was desired for the **response variable, a measure of disease progression one year after baseline.**
    
**Information about dataset found here:**
http://statweb.stanford.edu/~tibs/ftp/lars.pdf

##### Data Preparation

```julia
data = CSV.read("diabetes-data.csv")

println("Firts 5 Resposes for the Data:")
print(" ")
println(data[1:5,11])
println(" ")
println("Firts 5 Rows of Data:")
data[1:5,1:10]
```

```julia
println("Data Loaded!")
using StatsBase
rows = sample(1:442, 342, replace = false)
t_rows = [i for i in 1:size(data)[1]] 
filter!(x -> !(x in rows),t_rows)
train_data = data[rows,:]
test_data = data[t_rows,:]
println("Indexes for Train and Validation Dataset Created!")
```

```julia
dat = [x for x in zip(train_data[1],train_data[2],train_data[3],train_data[4],
            train_data[5],train_data[6],train_data[7],train_data[8],
            train_data[9],train_data[10], train_data[11])]
x_train = [x[1:10] for x in dat]
y_train = [x[11] for x in dat]

dat2 = [x for x in zip(test_data[1],test_data[2],test_data[3],test_data[4],
            test_data[5],test_data[6],test_data[7],test_data[8],
            test_data[9],test_data[10], test_data[11])]
x_test = [x[1:10] for x in dat2]
y_test = [x[11] for x in dat2]

println("Done organizing data for training and testing!")
```

##### Functions

```julia
function predict(w,x)
    """Linear combination of weights and input data"""
     x_new = push!([i for i in x],1.0)
    return w'x_new 
end

function total_MSE(w,x,y)
    """Mean Squared Error Cost Function"""
    sum = 0
    for i = 1:length(y)
        sum += ((y[i] - predict(w,x[i]))^2)
    end
    return sum/length(y)
end

function Derivative_MSE(w,x,y)
    """Derivative of the Mean Squared Error Cost Function"""
    xi = push!([i for i in x],1.0)
    sum = (predict(w,x) - y) * xi
    return (2 /length(y)) *sum
end

function update_weights(w,x,y,lr,j)
    """ Updating Weights to Minimize the Cost Function"""
    w_new = w - (lr * Derivative_MSE(w,x[j],y[j]))
    return w_new
end
```

##### Training the Model

```julia
list = []
n_features_bias = 10 + 1
w = randn(n_features_bias)
push!(list,total_MSE(w,x_train,y_train))

w = randn(n_features_bias)
w_hat = randn(n_features_bias)

for i = 1:700
    j = rand(1:342)
    w = update_weights(w,x_train,y_train,0.000006,j)
    
    if total_MSE(w,x_train,y_train) < total_MSE(w_hat,x_train,y_train)
        w_hat = copy(w)
    end
    push!(list,total_MSE(w,x_train,y_train))
end
```

```julia
println("---------------------------------")
println("Training Cost  |  Validation Cost")
println("  ",round(total_MSE(w_hat,x_train,y_train)),"       |       ",round(total_MSE(w_hat,x_test,y_test)))
println("---------------------------------")
plot(list)
plot!(title = "Loss - Diabetes Data", xlabel = "Iterations", ylabel = "Loss")
```

## Find Weights using Matrix Multiplication!

Algorithm :
* Set up matrix X with last columns being of the constant 1.
* Compute pseudo-inverse (pX) of matrix X.
* Calculate weights like so: W = pX*y

`If XTX is invertible, w = pX*y where pX = inv(XTX)*XT is the pseudo-inverse
of X. The resulting w is the unique optimal solution. If XTX is not
invertible, a pseudo-inverse can still be defined, but the solution will not be
unique`

```julia
X_matrix = convert(Matrix,train_data[:,1:10])
X_matrix = hcat(X_matrix,ones(342))
X_matrix_T = X_matrix'
xtx = X_matrix_T * X_matrix
xtx_inv = inv(xtx)
Xs = xtx_inv*X_matrix_T
W= Xs*y_train
print("These are the weights: \n")
print("----------------------------")
W
```

```julia
println("---------------------------------")
println("Training Cost  |  Validation Cost")
println("  ",round(total_MSE(W,x_train,y_train)),"       |       ",round(total_MSE(W,x_test,y_test)))
println("---------------------------------")
```

As we increase the number of features to our models, the MSE decreases. Thus, I have decided to use all features in the model.

```julia
train_cost = [3057.0,3083.0,3212.0 ,3214.0 ,3778.0 ,3784.0 ,3804.0 ,4047.0 ,5835.0,5835.0]    
valid_cost = [2284.0,2193.0 ,2372.0,2380.0,2787.0,2816.0,2745.0,3182.0,5330.0,5331.0]
plot(sort(train_cost, rev=true), label = "Train MSE")
plot!(sort(valid_cost, rev=true),label = "Validation MSE")
plot!(title = "MSE vs Number of Features", xlabel = "Number of Features", ylabel = "MSE")
```
